```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataHousePricing=pd.read_csv('house_pricing.csv')
dataHousePricing.head()
```

```python
dataHousePricing.info()
```

```python
sns.scatterplot(x=dataHousePricing["LotArea"],y=dataHousePricing["SalePrice"],data=dataHousePricing)
```

```python
filter = (dataHousePricing["LotArea"] < 20000) & (dataHousePricing["SalePrice"] < 500000)
newDHP = dataHousePricing.where(filter)
sns.scatterplot(x=newDHP["LotArea"],y=newDHP["SalePrice"],data=newDHP)
```

```python
sns.scatterplot(x=dataHousePricing["LotFrontage"],y=dataHousePricing["LotArea"],data=dataHousePricing)
```

```python
filter = (dataHousePricing["LotFrontage"] < 200) & (dataHousePricing["LotArea"] < 100000)
newDHP = dataHousePricing.where(filter)
sns.scatterplot(x=newDHP["LotFrontage"],y=newDHP["LotArea"],data=newDHP)
```
