```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataIBM=pd.read_csv('IBM_HR_ATTRITION.csv')
dataIBM.info()
```

```python
sns.displot(x="MonthlyIncome", data=dataIBM.where((dataIBM["MonthlyIncome"] <= 5000)), hue="Gender")
```

```python
newDataSet = dataIBM.replace({'Attrition': {'Yes': 1, 'No': 0}})
sns.lmplot(x="DistanceFromHome", y="Attrition", data=newDataSet)
```

```python
sns.lmplot(x="YearsWithCurrManager", y="Attrition", data=newDataSet)
```
