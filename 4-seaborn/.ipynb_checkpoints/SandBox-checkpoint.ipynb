```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataIBM=pd.read_csv('IBM_HR_ATTRITION.csv')
dataIBM["MonthlyRate"].head(100)
```

```python
dataIBM.info()
```

```python
result = sns.relplot(x="YearsAtCompany",y="MonthlyIncome", hue=(dataIBM["YearsAtCompany"] >= 20),data=dataIBM.sample(100))
legend = result._legend
legend.set_title(">= 20 years at company")
for t, l in zip(legend.texts,("No", "Yes")):
    t.set_text(l)
```
