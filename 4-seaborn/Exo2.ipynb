```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataSalesPredictions=pd.read_csv('sales_predictions.csv')
dataSalesPredictions.head()
```

```python
sns.relplot(x=dataSalesPredictions["item_price"],y=dataSalesPredictions["date"],data=dataSalesPredictions)
```

```python
newDSP = dataSalesPredictions.sample(50)
```

```python
sns.relplot(x=newDSP["item_price"],y=newDSP["date"],data=newDSP)
```

```python
a4_dims = (11.7, 8.27)
fig, ax = plt.subplots(figsize=a4_dims)
sns.scatterplot(x=newDSP["item_price"],y=newDSP["date"],ax=ax,data=newDSP)
```

```python
newDSP["date"] = pd.to_datetime(newDSP["date"])
sns.scatterplot(x="item_price",y="date",data=newDSP)
```
