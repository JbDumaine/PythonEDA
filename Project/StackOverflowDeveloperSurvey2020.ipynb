A look into the lives of over 64,000 Stack Overflow developers

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
pd.set_option('display.max_columns', 100)
pd.set_option('display.width', 1000)
df=pd.read_csv('survey_results_public.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
fig, ax = plt.subplots(figsize=(11.7, 8.27))
sns.countplot(data=df,y="Gender",ax=ax,order=df.Gender.value_counts().iloc[:4].index)
```

Conclusion : Les utilisateurs de stackoverflow sont majoritairement des hommes.

```python
sns.set_theme(style="darkgrid")
fig, ax = plt.subplots(figsize=(29.7, 21))
sns.countplot(data=df, y='Country', order=df.Country.value_counts().iloc[:10].index)
```

```python
sns.catplot(y="Gender", x="WorkWeekHrs", data=df, height=10, kind="bar")
```

```python
df2 = df.replace({'JobSat': {'Very satisfied': 5, 'Slightly satisfied': 4,'Neither satisfied nor dissatisfied':3,'Slightly dissatisfied':2,'Very dissatisfied':1}})
sns.catplot(y="Gender",x="JobSat", data=df2, height=10, kind="box")
```

```python
sns.catplot(y="Ethnicity",x="WorkWeekHrs", data=df2, height=10, kind="bar",order=df.Ethnicity.value_counts().iloc[:5].index)
```
