A look into the lives of over 64,000 Stack Overflow developers

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
pd.set_option('display.max_columns', 100)
pd.set_option('display.width', 1000)
df=pd.read_csv('survey_results_public.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
fig, ax = plt.subplots(figsize=(11.7, 8.27))
sns.countplot(data=df,y="Gender",ax=ax)
```

Conclusion : Les utilisateurs de stackoverflow sont majoritairement des hommes.

```python
sns.set_theme(style="darkgrid")
fig, ax = plt.subplots(figsize=(29.7, 21))
sns.countplot(data=df, y='Country', order=df.Country.value_counts().iloc[:3].index)
```
